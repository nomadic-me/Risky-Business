## Initial Imports

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced

# Import train_test_split from sklearn
from sklearn.model_selection import train_test_split

# Import the StandardScaler for Data Pre-Processing
from sklearn.preprocessing import StandardScaler

from imblearn.ensemble import BalancedRandomForestClassifier

# Import the Easy Ensemble Classifier (EEC)
from imblearn.ensemble import EasyEnsembleClassifier
```

## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

## Split the Data into Training and Testing

```python
# Create our features
X = df.copy()
X = X.drop(columns = 'loan_status')

# Create our target
y = df['loan_status']
```

### Encoding Data using `One-Hot Encoding`

```python
# Encoding Data using `One-Hot Encoding`
X = pd.get_dummies(X)
X.head()
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
# Create X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)

# Shape X_train
X_train.shape
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
brf.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
y_pred = brf.predict(X_test_scaled)
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix

pd.DataFrame(confusion_matrix(y_test, y_pred), 
            index = ["High Risk (Predicted Positive)", "Low Risk (Predicted Negative)"], 
            columns = ["Positive", "Negative"])
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

```python
# List the features sorted in descending order by feature importance
feature_importances = brf.feature_importances_ 
indices = np.argsort(feature_importances)[::-1]
for i in range(X.shape[1]):
    if len(X.columns.values[indices[i]])> 27: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t({feature_importances[indices[i]]})")
    elif len(X.columns.values[indices[i]])> 25: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t\t({feature_importances[indices[i]]})")
    elif len(X.columns.values[indices[i]])>24: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t({feature_importances[indices[i]]})")
    elif len(X.columns.values[indices[i]])>19: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t\t({feature_importances[indices[i]]})")
    elif len(X.columns.values[indices[i]])>11: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t\t\t({feature_importances[indices[i]]})")
    elif len(X.columns.values[indices[i]])>9: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t\t\t\t({feature_importances[indices[i]]})")
    elif len(X.columns.values[indices[i]])>8: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t\t\t\t({feature_importances[indices[i]]})")        
    elif len(X.columns.values[indices[i]])<4: 
        print(f"{i:02d} {X.columns.values[indices[i]]}:\t\t\t\t\t({feature_importances[indices[i]]})")    
```

### Easy Ensemble Classifier

```python
# Train the Classifier
eec = EasyEnsembleClassifier(n_estimators=100, random_state=1)
eec.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
y_pred = eec.predict(X_test_scaled)
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
pd.DataFrame(confusion_matrix(y_test, y_pred), 
                         index = ["High Risk (Predicted Positive)", "Low RIsk (Predicted Negative)"], 
                         columns = ["Positive", "Negative"])
```
